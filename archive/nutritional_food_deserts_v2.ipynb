## üß† Project Vision
This project aims to visualize the concept of **nutritional food deserts**, focusing on how access to healthy food varies by location. Inspired by "Banking Desert" studies, the goal is to show how lack of access to nutritious options may correlate with **health outcomes** such as **diabetes** and **high blood pressure**.

### ‚úÖ Core Objectives:
1. Analyze restaurants, convenience stores, and grocery stores to assess nutrition quality of food options.

3. Map access to nutritious food geographically by ZIP code.
4. Visualize health outcome data (e.g., diabetes prevalence) by area.
5. Study the correlation between poor health outcomes and nutritional access.

# Nutritional Score Formula and Explanation

## Formula

**Nutritional Score (NS)**:

\[
NS = \frac{
w_1 \cdot \text{normalize(calories)} + 
w_2 \cdot \text{normalize(protein\_g)} + 
w_3 \cdot \text{normalize(carbs\_g)} + 
w_4 \cdot \text{normalize(fat\_g)} + 
w_5 \cdot \text{normalize(cholesterol\_mg)} + 
w_6 \cdot \text{normalize(sodium\_mg)}
}{\text{total\_weight}}
\]

- `normalize()` scales each raw value to a 0‚Äì1 range based on estimated healthy maximums.
- Weights (`w1` to `w6`) prioritize different nutrients.
- The final score is between **0 (least healthy)** and **5 (most healthy)**.

---

## Explanation

| Nutrient         | Max / Reference Value | Source / Reasoning |
|------------------|------------------------|---------------------|
| Calories         | 1000 kcal              | Upper meal limit (CDC, USDA) |
| Protein (g)      | 50 g                   | Recommended daily intake (~50g; NIH, USDA) |
| Carbohydrates (g)| 150 g                  | WHO/FAO daily recommendations |
| Fat (g)          | 70 g                   | Upper daily limit (WHO) |
| Cholesterol (mg) | 300 mg                 | Upper limit per American Heart Association |
| Sodium (mg)      | 2400 mg                | Max recommended daily sodium intake (WHO) |

```python
def calculate_nutritional_score(calories, protein_g, carbs_g, fat_g, cholesterol_mg, sodium_mg):
    # Define max/reference values
    max_calories = 1000
    max_protein = 50
    max_carbs = 150
    max_fat = 70
    max_cholesterol = 300
    max_sodium = 2400
    
    # Weights for each component
    w_calories = -1  # Neutral or negative for high calories
    w_protein = 1
    w_carbs = 0.5
    w_fat = -0.5
    w_cholesterol = -1
    w_sodium = -1
    
    # Normalize
    norm_calories = max(0, min(1, calories / max_calories))
    norm_protein = max(0, min(1, protein_g / max_protein))
    norm_carbs = max(0, min(1, carbs_g / max_carbs))
    norm_fat = max(0, min(1, fat_g / max_fat))
    norm_cholesterol = max(0, min(1, cholesterol_mg / max_cholesterol))
    norm_sodium = max(0, min(1, sodium_mg / max_sodium))
    
    # Calculate score
    score = (
        w_calories * norm_calories +
        w_protein * norm_protein +
        w_carbs * norm_carbs +
        w_fat * norm_fat +
        w_cholesterol * norm_cholesterol +
        w_sodium * norm_sodium
    )
    
    # Shift and scale to 0-5
    final_score = max(0, min(5, (score + 5) / 2))
    return final_score
```
